```python
from WaterRocket import WaterRocket
```

### 1. Initialization

Add all informations about the rocket and the experimental environment like this

```python
myRocket = WaterRocket(
    bottle_volume= 2,         # 瓶子容积（单位：升）
    d_bottle= 8.9,            # 瓶子直径（单位：厘米）
    d_output= 0.9,            # 喷嘴直径（单位：厘米）
    m_empty_rocket= 0.5,      # 空瓶小火箭质量（单位：千克）
    Cx= 0.1,                  # 火箭的空气动力学系数
    tilt_angle= 89,           # 发射台的倾斜角度（单位：度）
    length_rampe= 22,         # 发射台长度（单位：厘米）
    initial_pressure= 10,     # 初始瓶内压力（单位：bar）
    initial_water_volume= 0.65,# 瓶内初始水量（单位：升）
    g=9.81,                   # 重力加速度（单位：m/s²）
    r=998,                    # 水的密度（单位：kg/m³）
    ra=1.2,                   # 空气的密度（单位：kg/m³）
    Patm=101325               # 大气压（单位：帕斯卡）
)
```

### 2. Simulate the flight

```python
data_flight = myRocket.create_df(save_as_CSV=False)
data_flight
```

### 3. Main informations about the flight

```python
myRocket.show_flight_infos(save_in_text=False)
```

### 4. Generate flights graphics

```python
myRocket.graphic_all(save_fig=False, show_figure=True)
```

### 5. Generate PDF report about the flight

```python
myRocket.createPDF(path_to_save_pdf="water_rocket_report.pdf", saveImgs=False, author="rocket")
```
